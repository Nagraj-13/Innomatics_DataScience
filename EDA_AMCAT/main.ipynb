```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df["Salary"], bins=20, edgecolor="black")
plt.xlabel("Salary")
plt.ylabel("Number of employees")
plt.title("Histogram of Salary")
plt.grid(axis="y")
plt.show()

plt.figure(figsize=(8, 6))
plt.boxplot(df['Salary'], vert=False)
plt.title('Box Plot of Salary')
plt.xlabel('Salary')
plt.show()


q1_S=df["Salary"].quantile(0.25)
q3_S=df["Salary"].quantile(0.75)
print(q1_S)
print(q3_S)
print(q3_S-q1_S)
```

The salary distribution appears to be right-skewed, meaning there are more employees towards the lower end of the salary range and fewer towards the higher end.

### Key statistics:
-Median: The median salary is around ₹300,000. This represents the value that separates the lower half of the employees from the higher half in terms of salary.
-First quartile (Q1): The first quartile is around ₹180,000. This means that 25% of the employees earn less than ₹180,000.
-Third quartile (Q3): The third quartile is around ₹370,000. This means that 75% of the employees earn less than ₹370,000.
-Interquartile range (IQR): The IQR is the difference between Q3 and Q1, which is around ₹190,000 in this case. This represents the range of salaries that contain the middle 50% of employees.

### Outliers:

There are several outliers on the right side of the plot, with salaries exceeding ₹1 million. It's important to investigate these outliers further to understand if they are valid data points or errors.


Now it can be observed the outliers lies in the right most part of the Q3 can be treated as rare cases, because as a fresher its highly difficult to get a salary package more than 15-20 LPA

```python
df.columns[2:]

selected_columns = ['Salary', '10percentage', '12percentage', 
        'collegeGPA', 'English', 'Logical', 'Quant', 'Domain',
       'ComputerProgramming', 'ComputerScience', 'conscientiousness',
       'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']

selected_df = df[selected_columns]
selected_df.head()
corr_matrix = selected_df.corr()

plt.figure(figsize=(8, 8), dpi=100)
# sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
sns.heatmap(selected_df.corr())
plt.title('Correlation Heatmap')
plt.show()
```

### General observations:

The heatmap displays the correlation coefficients between various numerical variables in your dataset, with "Salary" as the target variable.
The color scale ranges from blue (negative correlation) to yellow (positive correlation), with darker shades indicating stronger correlations.
The heatmap is symmetrical, as the correlation between X and Y is the same as the correlation between Y and X.


### Specific observations:

Academic performance: There seems to be a moderate positive correlation between "Salary" and "collegeGPA". However, the correlations with "10percentage" and "12percentage" are weaker.
Skills: Several skills show positive correlations with "Salary", including "ComputerProgramming", "Quant", and "Domain". Interestingly, "English" also shows a weak positive correlation.
Personality traits: The correlations between personality traits and "Salary" are mostly weak and mixed. "Conscientiousness" and "Openess_to_experience" show slightly positive trends, while others like "Extraversion" and "Neuroticism" have weak negative correlations.


### Potential interpretations:

Higher academic performance, particularly in higher education ("collegeGPA"), might be associated with higher salaries. However, earlier academic performance (represented by "10percentage" and "12percentage") seems to have a weaker influence.
Technical skills like "ComputerProgramming", "Quant", and "Domain" appear to be positively correlated with "Salary", suggesting their potential value in the job market. The positive correlation with "English" could be related to communication and soft skills required in many professions.
The weak and mixed correlations between personality traits and "Salary" indicate that personality might not be a sole determinant of income in this dataset. However, further analysis could explore potential interactions between different traits and their combined effect on salary.

```python

from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

chi2_stat, p_val, _, _ = chi2_contingency(contingency_table)
print("Chi-Squared Statistic:", chi2_stat)
print("P-value:", p_val)

alpha = 0.05
if p_val < alpha:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])


contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Specialization Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.legend().remove()  
# plt.legend(title='Specialization')
plt.tight_layout()
plt.show()

contingency_table.plot(kind='bar', figsize=(10, 6))
plt.title('Specialization Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend().remove()   
# plt.legend(title='Specialization')
plt.tight_layout()
plt.show()

```

It can be observed that by checking chi-square and the count plot the preference of the specialization depends on the gender

```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
# Read data and concatenate text
data = pd.read_excel("data.xlsx")
specializations = data['Specialization']
text = ' '.join(specializations.dropna())

# Create a mask
mask = np.array(Image.open("cloud_mask.png"))

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color=None, mode="RGBA", mask=mask).generate(text)

# Create a color gradient
image_colors = ImageColorGenerator(mask)

# Plot word cloud with gradient colors
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation='bilinear')
plt.title('Word Cloud of Specializations')
plt.axis('off')
plt.show()
```
